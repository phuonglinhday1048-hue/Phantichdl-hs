```
import pandas as pd
import numpy as np
```

```
#Vấn đề 1
column_names = [
    "Id", "Name", "Age", "Weight",
    "m0006", "m0612", "m1218",
    "f0006", "f0612", "f1218"
]

df = pd.read_csv(
    "heart_disease_dataset.csv",
    names=column_names
)

print(df.head())
```

```
#Vấn đề 2
# Xử lý lỗi: Các giá trị trong cột 'Name' không phải lúc nào cũng có 2 phần để tách.
# Ta sẽ tách tối đa 1 lần (n=1) và điền rỗng cho phần 'Lastname' nếu không có.
split_names = df['Name'].astype(str).str.split(n=1, expand=True)

df['Firstname'] = split_names[0]
# Gán 'Lastname' là phần thứ hai nếu có, nếu không thì là chuỗi rỗng
df['Lastname'] = split_names[1].fillna('') if split_names.shape[1] > 1 else ''

df = df.drop('Name', axis=1)
print (df)
```

```
#Vấn đề 3
def convert_weight(w):
    if pd.isna(w):
        return None

    w = str(w).strip().lower()

    if 'lbs' in w:
        return round(float(w.replace('lbs', '')) / 2.2, 2)

    elif 'kg' in w:
        return float(w.replace('kg', ''))

    return None

df['Weight'] = df['Weight'].apply(convert_weight)

print(df['Weight'].head())
```

```
#Vấn đề 4
df.dropna(how ="all", inplace=True)
print(df)
```

```
#Vấn đề 5
df = df.drop_duplicates(subset=['Firstname', "Lastname", "Age", "Weight"])
print(df)
```

```
#Vấn đề 6
df.Firstname.replace({r'[^\x00-\x7F]+':''}, regex=True, inplace=True)
df.Lastname.replace({r'[^\x00-\x7F]+':''}, regex=True, inplace=True)
print(df)
```

```
#Vấn đề 7
#Thống kê thông tin dữ liệu thiếu trên từng biến Age và Weight
print("Số lượng giá trị thiếu:")
print(df[['Age', 'Weight']].isnull().sum())
```

```
#Xóa các dòng thiếu cả Age và Weight
df = df.drop(
    df[
        (df['Age'].isnull()) &
        (df['Weight'].isnull())
    ].index
)
#Thay thế giá trị thiếu của Age bằng giá trị trung bình
df['Weight'] = (
    df['Weight']
    .astype(str)
    .str.extract(r'(\d+\.?\d*)')[0]
    .astype(float)
)

df['Weight'].fillna(df['Weight'].mean(), inplace=True)
df[['Age', 'Weight']].isnull().sum()
```

```
#Vấn đề 8: Một cột chứa quá nhiều thông tin cần được phân rã
#Chuyển dữ liệu từ Wide sang Long

# First, flatten the MultiIndex columns of df, which were created due to incorrect data loading.
# From the kernel state, the first level of the MultiIndex columns contains the intended names.
if isinstance(df.columns, pd.MultiIndex):
    df.columns = df.columns.get_level_values(0)

# Now, ensure 'Id' (capitalized) is used, as 'id' (lowercase) does not exist.
df_long = df.melt(
    id_vars=['Id'], # Changed from 'id' to 'Id'
    var_name='Code',
    value_name='BloodPressure'
)
#Tách giới tính
df_long['Gender'] = df_long['Code'].str[0]
#Tách nhóm tuổi
df_long['AgeGroup'] = df_long['Code'].str[1:]
#Chuyển ký hiệu giới tính thành tên
df_long['Gender'] = df_long['Gender'].replace({
    'm': 'Male',
    'f': 'Female'
})
print(df_long.head())
```
