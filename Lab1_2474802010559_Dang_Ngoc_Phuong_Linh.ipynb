```python
import pandas as pd
import numpy as np
```

```python
#in 10 hàng dữ liệu ra
df = pd.read_csv('processed_dulieuxettuyendaihoc.csv',index_col='STT')
df.head(10)
```

```python
#in ra 10 dòng dữ liệu cuối cùng
df.tail(10)
```

```python
#Thống kê dữ liệu thiếu cho cột dân tộc và hiệu chỉnh dữ liệu thiếu như sau: Mặc định thiếu thì điền giá trị 0.
df["DT"].fillna(0, inplace = True)
df
```

```python
#Thống kê dữ liệu thiếu cho biến T1 và hiệu chỉnh dữ liệu, lưu ý việc thay thế dữ liệu thiếu sử dụng phương pháp Mean.
df["T1"].fillna(df["T1"].mean(), inplace = True)
df
```

```python
#Hãy thực hiện xử lý lần lượt tất cả dữ liệu thiếu cho các biến về điểm số còn lại.
Cot_diem = ['T1', 'L1', 'H1', 'S1', 'V1', 'X1', 'D1', 'N1',
            'T2', 'L2', 'H2', 'S2', 'V2', 'X2', 'D2', 'N2',
            'T6', 'L6', 'H6', 'S6', 'V6', 'X6', 'D6', 'N6',
            'DH1', 'DH2', 'DH3']

for col in Cot_diem:
    if col in df.columns:
        df[col] = df[col].fillna(df[col].mean())
df
```

```python
#Tạo các biến TBM1, TBM2, TBM3 tương ứng với trung bình môn của các năm lớp 10, 11 và 12.
df['TBM1'] = (df['T1']*2 + df['L1'] + df['H1'] + df['S1'] + df['V1']) + df['X1'] + df ['D1'] + df ['N1']/10
df['TBM2'] = (df['T2']*2 + df['L2'] + df['H2'] + df['S2'] + df['V2']) + df['X2'] + df ['D2'] + df ['N2']/10
df['TBM3'] = (df['T6']*2 + df['L6'] + df['H6'] + df['S6'] + df['V6']) + df['X6'] + df ['D6'] + df ['N6']/10
df
```

```python
#Tạo các biến xếp loại XL1, XL2 và XL3 dựa trên TBM1,TBM2 và TBM3 cho từng năm lớp 10, 11, 12 như sau:
#Nhỏ hơn 5.0 xếp loại: yếu (kí hiệu là Y)
#Từ 5.0 đến dưới 6.5: trung bình (kí hiệu là TB)
#Từ 6.5 đến dưới 8.0: khá (kí hiệu là K)
#Từ 8.0 đến dưới 9.0: giỏi (kí hiệu là G)
#Từ 9.0 trở lên: xuất sắc (kí hiệu là XS)
XL = ['Y', 'TB', 'K', 'G', 'XS']
diems = [0, 5.0, 6.5, 8.0, 9.0, 10.0]
df['XL1'] = pd.cut(df['TBM1'], bins=diems, labels=XL, right=False)
df['XL2'] = pd.cut(df['TBM2'], bins=diems, labels=XL, right=False)
df['XL3'] = pd.cut(df['TBM3'], bins=diems, labels=XL, right=False)
```
