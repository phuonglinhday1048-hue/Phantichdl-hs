```python
#LAB 02: BÀI THỰC HÀNH TRÌNH BÀY DỮ LIỆU
#Phần 1: Thống kê dữ liệu
#1. Hãy sắp xếp dữ liệu điểm DH1 theo thứ tự tăng dần
import pandas as pd

# Đọc dữ liệu
df = pd.read_csv("processed_dulieuxettuyendaihoc.csv")

# Sắp xếp DH1 tăng dần
df_sorted_dh1 = df.sort_values(by="DH1")

# In kết quả
print(df_sorted_dh1[["DH1"]])
```

```python
#3. Hãy tạo pivot-table để thống kê các giá trị count, sum, mean, median, min max, std, Q1, Q2 và Q3 của DH1 theo KT
# Hàm tính Q1, Q2, Q3
def Q1(x):
    return x.quantile(0.25)

def Q2(x):
    return x.quantile(0.50)

def Q3(x):
    return x.quantile(0.75)

# Pivot-table
pivot_KT = pd.pivot_table(
    df,
    values="DH1",
    index="KT",
    aggfunc=[
        "count",
        "sum",
        "mean",
        "median",
        "min",
        "max",
        "std",
        Q1,
        Q2,
        Q3
    ]
)

print(pivot_KT)
```

```python
#5. Hãy tạo pivot-table để thống kê các giá trị count, sum, mean, median, min, max, std Q1, Q2 và Q3 của DH1 theo KT, KV và DT
pivot_KT_KV_DT = pd.pivot_table(
    df,
    values="DH1",
    index=["KT", "KV", "DT"],
    aggfunc=[
        "count",
        "sum",
        "mean",
        "median",
        "min",
        "max",
        "std",
        Q1,
        Q2,
        Q3
    ]
)

print(pivot_KT_KV_DT)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('processed_dulieuxettuyendaihoc.csv')
```

2. Hãy tạo biến phân loại (phanlopt1) cho môn toán (T1) như sau:
a. Từ 0 đến dưới 5 = kém (ký hiệu “k”)
b. Từ 5 đến dưới 7 = trung bình (ký hiệu “tb”)
c. Từ 7 đến dưới 8 = khá (ký hiệu “k”)
d. Từ 8 trở lên = giỏi (ký hiệu “g”)

```python
def phan_loai_t1(diem):
    if diem < 5.0:
        return 'k'
    elif diem < 7.0:
        return 'tb'
    elif diem < 8.0:
        return 'kh'
    else:
        return 'g'

df['phanloait1'] = df['T1'].apply(phan_loai_t1)
print("Đã tạo thành công biến phanloait1.")
df[['T1', 'phanloait1']].head()
```

3. Lập bảng tần số cho biến phanloait1

```python
bang_tan_so_pl1 = df['phanloait1'].value_counts()

print("--- BẢNG TẦN SỐ BIẾN PHANLOAIT1 ---")
print(pd.DataFrame({'Tần số': bang_tan_so_pl1}))
```

4. Vẽ biểu đồ đường Multiple Line cho biến T1 được phân loại bởi biến phanloait1

```python
pivot_line = pd.crosstab(df['T1'], df['phanloait1'])

pivot_line = pivot_line.reindex(sorted(df['T1'].unique())).fillna(0)

plt.figure(figsize=(10, 6))
colors = {'k': 'green', 'tb': 'purple', 'kh': 'orange', 'g': 'blue'}
labels = {'k': 'Kém (k)', 'tb': 'Trung bình (tb)', 'kh': 'Khá (kh)', 'g': 'Giỏi (g)'}

for col in pivot_line.columns:
    if col in colors:
        plt.plot(pivot_line.index, pivot_line[col], label=labels[col], color=colors[col], linewidth=1.5)

plt.title('Biểu đồ đường Multiple Line cho biến T1 phân loại bởi phanloait1')
plt.xlabel('T1')
plt.ylabel('Count')
plt.legend(title='Phanloait1')
plt.grid(True, axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.savefig('multiple_line_t1.png')
plt.show()
```

5. Vẽ biểu đồ Drop-line cho biến T1 được phân loại bởi biến phanloait1

```python
df_counts = df.groupby(['T1', 'phanloait1']).size().reset_index(name='Count')

plt.figure(figsize=(12, 6))
colors = {'k': 'green', 'tb': 'purple', 'kh': 'orange', 'g': 'blue'}
labels = {'k': 'Kém (k)', 'tb': 'Trung bình (tb)', 'kh': 'Khá (kh)', 'g': 'Giỏi (g)'}

for idx, row in df_counts.iterrows():
    plt.vlines(x=row['T1'], ymin=0, ymax=row['Count'], color='black', linewidth=1, alpha=0.7)
    plt.scatter(row['T1'], row['Count'], color=colors[row['phanloait1']], s=25, zorder=3)

for key, value in colors.items():
    plt.scatter([], [], color=value, label=labels[key])

plt.title('Biểu đồ Drop-line cho biến T1 được phân loại bởi phanloait1')
plt.xlabel('T1')
plt.ylabel('Count')
plt.ylim(0, df_counts['Count'].max() + 0.5)
plt.legend(title='Phanloait1')
plt.grid(True, axis='y', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.savefig('dropline_t1.png')
plt.show()
```

Phần 5: Mô tả dữ liệu và khảo sát dạng phân phối
1. Hãy mô tả và khảo sát phân phối cho biến T1
Gợi ý:
- Mô tả độ tập trung và phân tán của dữ liệu T1
- Vẽ biểu đồ Box-Plot và xác định các 10 đại lượng trong biểu đồ đó
- Mô tả hình dáng lệch của phân phối T1 dựa vào các đại lượng hướng tâm
- Vẽ biểu đồ Histogram biểu thị hình dáng phân phối
- Mô tả các đặc trưng của phân phối, mức độ lệch và mức độ nhọn
- Kiểm chứng phân phối chuẩn QQ-Plot
- Nhận xét và đánh giá về phân phối của T1

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

# Đọc dữ liệu
df = pd.read_csv("processed_dulieuxettuyendaihoc.csv")

# Thống kê mô tả
print(df["T1"].describe())

# Các đại lượng hướng tâm
print("Mean =", df["T1"].mean())
print("Median =", df["T1"].median())
print("Mode =")
print(df["T1"].mode())

# Độ phân tán
print("Variance =", df["T1"].var())
print("Standard Deviation =", df["T1"].std())
print("Min =", df["T1"].min())
print("Max =", df["T1"].max())

# Tính các đại lượng Boxplot

Q1 = df["T1"].quantile(0.25)
Q2 = df["T1"].quantile(0.50)
Q3 = df["T1"].quantile(0.75)

IQR = Q3 - Q1

Lower = Q1 - 1.5 * IQR
Upper = Q3 + 1.5 * IQR

print("Min =", df["T1"].min())
print("Q1 =", Q1)
print("Median =", Q2)
print("Q3 =", Q3)
print("Max =", df["T1"].max())
print("IQR =", IQR)
print("Lower Bound =", Lower)
print("Upper Bound =", Upper)

# Outlier
outlier = df[(df["T1"] < Lower) | (df["T1"] > Upper)]

print("Số lượng Outlier =", len(outlier))

# Boxplot
sns.boxplot(x=df["T1"])

plt.title("Boxplot của T1")

plt.show()

# So sánh Mean và Median

mean_T1 = df["T1"].mean()
median_T1 = df["T1"].median()

print("Mean =", mean_T1)
print("Median =", median_T1)

if mean_T1 > median_T1:
    print("Phân phối lệch phải")

elif mean_T1 < median_T1:
    print("Phân phối lệch trái")

else:
    print("Phân phối đối xứng")

# Histogram
df["T1"].hist(bins=10)

plt.title("Histogram của T1")
plt.xlabel("T1")
plt.ylabel("Tần số")

plt.show()

# QQ Plot
stats.probplot(df["T1"], dist="norm", plot=plt)

plt.title("QQ-Plot của T1")

plt.show()

# Độ lệch
print("Skewness =", df["T1"].skew())

# Độ nhọn
print("Kurtosis =", df["T1"].kurt())

# QQ Plot
stats.probplot(df["T1"], dist="norm", plot=plt)

plt.title("QQ-Plot của T1")

plt.show()

print("""
Nhận xét:
- Dựa vào Histogram và QQ-Plot để đánh giá phân phối.
- Nếu các điểm trên QQ-Plot gần đường thẳng thì dữ liệu gần phân phối chuẩn.
- Skewness > 0: lệch phải.
- Skewness < 0: lệch trái.
- Kurtosis lớn: phân phối nhọn.
- Kurtosis nhỏ: phân phối bẹt.
""")
```

2. Hãy mô tả và khảo sát phân phối cho biến T1 trên từng nhóm phân lớp (phanlopT1)
Gợi ý
- Trực quan hóa biểu đồ Box-plot, histogram và QQ-plot theo phân nhóm là giá trị
của ‘phanlopT1’.

```python
# Tạo cột phân lớp T1

def phanlop(x):

    if x < 5:
        return "Y"

    elif x < 6.5:
        return "TB"

    elif x < 8:
        return "K"

    else:
        return "G"

df["phanlopT1"] = df["T1"].apply(phanlop)

print(df[["T1", "phanlopT1"]].head())

# Kiểm tra nhóm phân lớp
print(df["phanlopT1"].value_counts())

sns.boxplot(x="phanlopT1", y="T1", data=df)

plt.title("Boxplot T1 theo phanlopT1")

plt.show()

# Histogram theo nhóm
sns.histplot(data=df, x="T1", hue="phanlopT1", kde=True)

plt.title("Histogram T1 theo phanlopT1")

plt.show()

# Danh sách nhóm
nhom = df["phanlopT1"].unique()

for i in nhom:

    du_lieu = df[df["phanlopT1"] == i]

    stats.probplot(du_lieu["T1"], dist="norm", plot=plt)

    plt.title(f"QQ-Plot nhóm {i}")

    plt.show()
```

3. Hãy khảo sát tương quan giữa biến DH1 theo biến T1
Gợi ý
- Nhận xét giá trị Covariance hoặc Correlation
- Vẽ biểu đồ Scatter thể hiện liên hệ của biến phụ thuộc DH1 theo biến độc lập T1

```python
# Covariance
cov = df["DH1"].cov(df["T1"])

print("Covariance =", cov)

# Correlation
corr = df["DH1"].corr(df["T1"])

print("Correlation =", corr)

plt.scatter(df["T1"], df["DH1"])

plt.title("Scatter Plot giữa T1 và DH1")
plt.xlabel("T1")
plt.ylabel("DH1")

plt.show()
```

4. Hãy khảo sát tương quan giữa biến DH1 theo biến T1 trên từng nhóm khu vực

```python
# Scatter theo khu vực
sns.scatterplot(data=df, x="T1", y="DH1", hue="KV")

plt.title("DH1 và T1 theo khu vực")

plt.show()

# Correlation theo khu vực

for i in df["KV"].unique():

    du_lieu = df[df["KV"] == i]

    corr = du_lieu["DH1"].corr(du_lieu["T1"])

    print("KV =", i)
    print("Correlation =", corr)
    print()
```
